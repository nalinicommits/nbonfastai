# Build and train model on KDEF Dataset

## Download / Prepare dataset

```python
#set training path
trainingpath = path/'kdef'
print(trainingpath)
```

Images are already downloaded into specific folders under the training path. Only need to verify them.

```python
for i in range(len(emoclasses)):
    dest = trainingpath/emoclasses[i]
    print("Destination folder set to", dest)
    print("\nVerifying images now..")
    verify_images(dest, delete=True, max_workers=8)
```

Creating image databunch on this dataset now...

```python
np.random.seed(42)
kdefdata = ImageDataBunch.from_folder(trainingpath, train=".", valid_pct=0.2, 
                                      ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
kdefdata.classes, kdefdata.c, len(kdefdata.train_ds), len(kdefdata.valid_ds)
```

```python
kdefdata.show_batch(rows=5, figsize=(7,8))
```

## Train model on kdef dataset

```python
learnkdef = create_cnn(kdefdata, models.resnet34, metrics=error_rate)
```

```python
learnkdef.fit_one_cycle(8)
```

First stage done and saved. Now let us try to improve it by adjusting learning rates.

```python
learnkdef.unfreeze()
learnkdef.lr_find()
learnkdef.recorder.plot()
```

```python
learnkdef.fit_one_cycle(10, max_lr=slice(1e-6,1e-4))
```

```python
learnkdef.save('kdef-stage1')
```

Unfreeze and retrain with different epochs and learning rates. 
No need to do this if earlier stages yielded low error rates. Just save and load that stage later.

```python
learnkdef.unfreeze()
learnkdef.lr_find()
learnkdef.recorder.plot()
```

```python
learnkdef.fit_one_cycle(10, max_lr=slice(1e-6,1e-3))
```

```python
learnkdef.unfreeze()
learnkdef.lr_find()
learnkdef.recorder.plot()
```

```python
learnkdef.fit_one_cycle(10, max_lr=slice(1e-5,1e-4))
```

```python
learnkdef.save('kdef-stage2-errPt07')
```

## Interpret results from KDEF dataset

```python
learnkdef.load('kdef-stage2-errPt07')
interp = ClassificationInterpretation.from_learner(learnkdef)
interp.plot_confusion_matrix()
```

```python
interp.most_confused(min_val=2)
```

So with approx 25 mins of training, the model has a 7.5% error rate and is mostly confused between afraid and surprised. And TL is less than VL as it should be. Let's view the images to see if its confusing to us too.

```python
interp.plot_top_losses(9, figsize=(15,11))
```

Ok, so the most confused look pretty difficult to sort. I'm going to train some more to see if that helps.

```python
#determine lr slice first
learnkdef.unfreeze()
learnkdef.lr_find()
learnkdef.recorder.plot()
```

```python
learnkdef.fit_one_cycle(30, max_lr=slice(1e-7,1e-4))
```

```python
learnkdef.save('kdef-stage3-errPt05tlPt08vlPt16')
```
